```python
import sqlite3
import json

# Connect to the SQLite database (or create it if it doesn't exist)
conn = sqlite3.connect('database.db')
cursor = conn.cursor()

 # Create tables
 
 # create authors table
cursor.execute('''
               create table if not exists authors (
                    id text primary key,
                    display_name text,
                    works_count integer,
                    cited_by_count integer,
                    predicted_gender text
               )
''')

# create x-concepts table
cursor.execute('''
               create table if not exists x_concepts (
                    id text primary key,
                    x_concepts text
               )
''')


# create cited_by_year table
cursor.execute('''
CREATE TABLE IF NOT EXISTS cited_by_year (
    id TEXT,
    year INTEGER,
    works_count INTEGER,
    cited_by_count INTEGER
)
''')

# create works table
cursor.execute('''
CREATE TABLE IF NOT EXISTS works (
  id TEXT PRIMARY KEY,
  title TEXT,
  


# Function to insert data into cited_by_year table


def insert_cited_by_year(data):
    cursor.execute('''
    INSERT INTO cited_by_year (id, year, works_count, cited_by_count)
    VALUES (?, ?, ?, ?)
    ''', (data['id'], data['year'], data['works_count'], data['cited_by_count']))

# Function to insert data into authors table


def insert_authors(data):
    cursor.execute('''
    INSERT INTO authors (id, display_name, works_count, cited_by_count, predicted_gender)
    VALUES (?, ?, ?, ?, ?)
    ''', (data['id'], data['display_name'], data['works_count'], data['cited_by_count'], data['predicted_gender']))


# Read and insert data from cited_by_year.jsonl
with open('/Users/jukatz/Documents/GitHub/Project-3/data/jsonl/cited_by_year.jsonl', 'r') as file:
    for line in file:
        data = json.loads(line)
        insert_cited_by_year(data)

# Read and insert data from authors.jsonl
with open('/Users/jukatz/Documents/GitHub/Project-3/data/jsonl/authors.jsonl', 'r') as file:
    for line in file:
        data = json.loads(line)
        insert_authors(data)

# Commit the transaction and close the connection
conn.commit()
conn.close()
```
