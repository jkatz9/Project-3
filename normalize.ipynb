```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker

# Define the base class for declarative models
Base = declarative_base()

# Define a sample table as a Python class


class Author(Base):
    __tablename__ = 'authors'
    id = Column(String, primary_key=True)
    display_name = Column(String)

    name = Column(String)
    age = Column(Integer)


# Replace with your actual database URL
DATABASE_URL = "postgresql://jukatz@localhost/project3"

# Create an engine that stores data in the local PostgreSQL database
engine = create_engine(DATABASE_URL)

# Create all tables in the database
Base.metadata.create_all(engine)
```

```python
import pickle
import pandas as pd
import glob

# Create an empty DataFrame to store the results
df = pd.DataFrame()

# Use glob to find all files matching the pattern
file_pattern = 'data/gender_predictions_*.pkl'
file_list = glob.glob(file_pattern)

# Iterate over the list of files
for file_path in file_list:
    # Open and load each pickle file
    with open(file_path, 'rb') as file:
        data = pickle.load(file)

    # Convert the loaded data to a DataFrame
    # Assuming data is in a format that can be directly converted to a DataFrame
    temp_df = pd.DataFrame(data)

    # Append the data to the main DataFrame
    df = pd.concat([df, temp_df], ignore_index=True)

# Now `df` contains the combined data from all pickle files
print(df)
```
